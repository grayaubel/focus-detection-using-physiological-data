```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.DataFrame({
    'Window': ['30s', '60s', '120s', '180s'],
    'Precision': [0.81, 0.77, 0.80, 0.75],
    'Recall': [0.92, 0.90, 0.92, 0.80],
    'F1-score': [0.83, 0.80, 0.84, 0.77]
})
```

```python
colors = ['#708090', '#DCD6F7', '#6A0DAD', '#A6B1E1']
```

```python
df_plot = df.set_index('Window').T
```

```python
df_plot.plot(kind='bar', color=colors, width=0.7)
plt.title("Performance Comparison Across Metrics and Window Sizes")
plt.ylabel("Score")
plt.ylim(0.7, 1)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.legend(title='Window Size')
plt.tight_layout()
plt.show()
```
